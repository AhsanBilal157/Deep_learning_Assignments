# **Problem Statement::**

A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts.

They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:

Which variables are significant in predicting the price of a car How well those variables describe the price of a car Based on various market surveys, the consulting firm has gathered a large data set of different types of cars across the America market.

# **task::¶**
We are required to model the price of cars with the available independent variables. It will be used by the management to understand how exactly the prices vary with the independent variables. They can accordingly manipulate the design of the cars, the business strategy etc. to meet certain price levels. Further, the model will be a good way for management to understand the pricing dynamics of a new market.

```python
#encode_data = pd.get_dummies(car_price_pridiction,columns=['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber'])
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
car_price_pridiction['fueltype'] = labelencoder.fit_transform(car_price_pridiction['fueltype'])
car_price_pridiction['aspiration'] = labelencoder.fit_transform(car_price_pridiction['aspiration'])
car_price_pridiction['carbody'] = labelencoder.fit_transform(car_price_pridiction['carbody'])
car_price_pridiction['drivewheel'] = labelencoder.fit_transform(car_price_pridiction['drivewheel'])
car_price_pridiction['enginelocation'] = labelencoder.fit_transform(car_price_pridiction['enginelocation'])
car_price_pridiction['fuelsystem'] = labelencoder.fit_transform(car_price_pridiction['enginelocation'])
```

```python
car_price_pridiction["doornumber"] = car_price_pridiction["doornumber"].map({'two':2,'four':4})
car_price_pridiction['doornumber']
```

```python
car_price_pridiction.head()
```

```python
data1 = car_price_pridiction[['horsepower','boreratio','enginesize','curbweight','carwidth','carlength',
              'wheelbase','drivewheel','price','enginelocation']]
```

```python
data_x_input = data1.loc[:, data1.columns != 'price']
data_y_input = data1.loc[:,data1.columns == 'price']
from sklearn.model_selection import train_test_split

train_data,test_data,train_targets,test_targets =  train_test_split(data_x_input,data_y_input , test_size = 0.30 , random_state = 42)
```

```python
train_data
```

```python
mean= train_data.mean(axis=0)
train_data -= mean
std=train_data.std(axis=0)
train_data /= std

test_data -= mean
test_data /=std
```

```python
mean_label = train_targets.mean()
train_targets -= mean_label
std_label = train_targets.std()
train_targets /= std_label
test_targets -= mean_label
test_targets /= std_label
```

```python
train_targets
```

```python
test = np.array(test_data.iloc[:])
```

```python
train = np.array(train_data.iloc[:])
```

```python
test_l= np.array(test_targets.astype('float32'))
```

```python
train_l= np.array(train_targets.astype('float32'))
```

```python
from keras import models
from keras import layers

def build_model():
  model = models.Sequential()
  model.add(layers.Dense(10, input_shape=(train.shape[1],)))
  model.add(layers.Dense(8,))
  model.add(layers.Dense(6,activation = 'tanh'))
  model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
  return model
```

```python
import numpy as np
k=4
num_val_samples = len(train) // k
num_epochs = 100
all_scores = []
for i in range(k):
  print('processing fold #', i)
  val_data = train[i * num_val_samples: (i + 1) * num_val_samples]
  val_targets = train_l[i * num_val_samples: (i + 1) * num_val_samples]
  partial_train_data = np.concatenate(
  [train[:i * num_val_samples],
  train[(i + 1) * num_val_samples:]],
  axis=0)
  partial_train_l = np.concatenate(
  [train_l[:i * num_val_samples],
  train_l[(i + 1) * num_val_samples:]],
  axis=0)
  model = build_model()
  model.fit(partial_train_data, partial_train_l,
          epochs=num_epochs, batch_size=1, verbose=1)
  val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=1)
  all_scores.append(val_mae)
```

```python
all_scores
```

```python
model = build_model()
model.fit(train, train_l,
epochs=80, batch_size=1, verbose=0)
test_mse_score, test_mae_score = model.evaluate(test, test_l)
```
